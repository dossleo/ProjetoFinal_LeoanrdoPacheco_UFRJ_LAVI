```python
defeitos = [
        'normal',

        'desalinhamento_horizontal_baixo',
        'desalinhamento_horizontal_mÃ©dio',
        'desalinhamento_horizontal_mÃ©dio',
        'desalinhamento_horizontal_alto',

        'desalinhamento_vertical_baixo',
        'desalinhamento_vertical_baixo',
        'desalinhamento_vertical_mÃ©dio',
        'desalinhamento_vertical_mÃ©dio',
        'desalinhamento_vertical_alto',
        'desalinhamento_vertical_alto',

        'desbalanceamento_baixo',
        'desbalanceamento_baixo',
        'desbalanceamento_medio',
        'desbalanceamento_medio',
        'desbalanceamento_alto',
        'desbalanceamento_alto',
        'desbalanceamento_alto',

        # 'ball_fault_baixo',
        'ball_fault_baixo',
        'ball_fault_medio',
        'ball_fault_alto',

        # 'cage_fault_baixo',
        # 'cage_fault_baixo',
        # 'cage_fault_medio',
        # 'cage_fault_alto',

        # 'outer_race_baixo',
        'outer_race_baixo',
        'outer_race_medio',
        'outer_race_alto',

        # 'ball_fault_baixo',
        'ball_fault_baixo',
        'ball_fault_medio',
        'ball_fault_alto',

        # 'cage_fault_baixo',
        # 'cage_fault_baixo',
        # 'cage_fault_medio',
        # 'cage_fault_alto',

        # 'outer_race_baixo',
        'outer_race_baixo',
        'outer_race_medio',
        'outer_race_alto',
]
```

```python

for defeito in defeitos:
# Adicionando uma nova coluna com valores condicionais
    df[defeito] = df.apply(lambda row: 1 if row['defeito'] == defeito else 0, axis=1)

print(df)
```

```python
x = df[models.x]
y = df[defeitos]
```

```python
x.dtypes
```

```python
# Importando as bibliotecas necessÃ¡rias
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Separando os dados em conjunto de treinamento e conjunto de teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Criando o modelo de classificaÃ§Ã£o com o algoritmo Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Treinando o modelo com os dados de treinamento
clf.fit(X_train, y_train)

# Fazendo previsÃµes com os dados de teste
y_pred = clf.predict(X_test)

# Calculando a acurÃ¡cia do modelo
accuracy = accuracy_score(y_test, y_pred)

# Imprimindo a acurÃ¡cia
print('AcurÃ¡cia:', accuracy)
```
